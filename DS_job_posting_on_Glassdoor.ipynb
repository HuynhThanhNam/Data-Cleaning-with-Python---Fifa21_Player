```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Uncleaned_DS_jobs.csv', low_memory=False)
df.index = df.index + 1
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 1000)
df.drop('index',axis = 1, inplace = True)
df.head()
```

### Remove duplicated

```python
#check duplicated
print(df[df.duplicated()].shape)
```

```python
#Data duplicated
df[df.duplicated()]
```

```python
#Drop duplicated
df.drop_duplicates(inplace = True)
df.reset_index(inplace=True)
df.info()
```

### Remove numbers from Company Name column

```python
df['Company Name']
```

```python
#remove number after '\n' (=get name of company before \n)
df1 = df['Company Name'].str.split('\n').str[0]
df.update(df1)
df.head()
```

### Create State Column from Location column

```python
#extract State from Location
df['State'] = df['Location'].str.split(',').str[-1].str.strip(' ')
df.head()
```

```python
df['State'].value_counts().to_frame()
```

```python
def rename_state(rs):
    if rs == 'United States':
        return 'U.S'
    elif rs == 'New Jersey':
        return 'NJ'
    elif rs == 'Texas':
        return 'TX'
    elif rs == 'California':
        return 'CA'
    elif rs =='Utah':
        return 'UT'
    else:
        return rs

df['State'] = df['State'].apply(rename_state)
```

```python
df['State'].value_counts().to_frame()
```

### Cleaning Salary Estimate column

```python
df['Salary Estimate']
```

```python
#2 formats:
#79K−131K (Glassdoor est.)
#145K−225K(Employer est.)

#remove (Glassdoor est.),(Employer est.)
dff = df['Salary Estimate'].str.strip(' (Glassdoor est.),(Employer est.)')
df.update(dff)
df['Salary Estimate']
```

```python
#add min salary , max salary and average salary
df['Min_Salary'] =  df['Salary Estimate'].str.split('-').str[0].str.strip('$K').astype(int)
df['Max_Salary'] =  df['Salary Estimate'].str.split('-').str[1].str.strip('$K').astype(int)
df['AVG_Salary'] = df[['Min_Salary','Max_Salary']].mean(axis = 1)

#Convert from 'Salary Estimate' column to format: 'Min_salary - Max_Salary'
df['Salary Estimate'] = df['Min_Salary'].astype(str) + '-' + df['Max_Salary'].astype(str)
df.rename({'Salary Estimate':'Salary Estimate(K$)'}, axis = 1, inplace = True)
df.head()
```

### Calculate Age of Company

```python
df['Company_Age'] = df['Founded'].apply(lambda x: 'unknown' if x == -1 else 2023 - x)
df['Company_Age']
```

### Extracting key skills mentioned in the job description

```python
def key_skills(key_skill):
    df[key_skill] = df['Job Description'].apply(lambda x: 1 if key_skill in x.lower() else 0)
    
key_skills('python')
key_skills('excel')
key_skills('hadoop')
key_skills('spark')
key_skills('aws')
key_skills('tableau')
key_skills('big_data')
df.head()
```

### Create seniority column from Job Title Column

```python
df['Job Title'].value_counts().to_frame()
```

```python
def seniority(job_title):
    job_title=job_title.lower()
    senior=['sr','senior','lead','principal','vp','vice president','director']
    for i in senior:
        if i in job_title:
            return "senior"
    if "jr" in job_title:
        return "junior"
    return "all level"

df['Seniority'] = df['Job Title'].apply(seniority)
df['Seniority'].value_counts().to_frame()
```

```python
#select rows where Seniority = 'senior'
df[df['Seniority'] == 'senior']
```
